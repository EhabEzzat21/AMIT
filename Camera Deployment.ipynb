```python
import cv2
import numpy as np
from keras.models import model_from_json

# Load the trained model
with open('best_model.json', 'r') as json_file:
    loaded_model_json = json_file.read()

loaded_model = model_from_json(loaded_model_json)
loaded_model.load_weights('model_weights.h5')

# Define the emotion dictionary
emotion_dict = {0: 'Angry', 1: 'Happy', 2: 'Neutral', 3: 'Sad', 4: 'Startle'}

# Open the webcam
cap = cv2.VideoCapture(0)  # 0 is the default camera

# Check if the webcam is opened correctly
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

# Read frames from the webcam
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Preprocess the frame
    resized_frame = cv2.resize(frame, (48, 48))
    rgb_frame = cv2.cvtColor(resized_frame, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB
    normalized_frame = rgb_frame / 255.0
    input_data = np.expand_dims(normalized_frame, axis=0)

    # Perform prediction
    prediction = loaded_model.predict(input_data)
    emotion_label = emotion_dict[np.argmax(prediction)]

    # Display emotion label on the frame
    cv2.putText(frame, emotion_label, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    # Display the frame
    cv2.imshow('Webcam Emotion Detection', frame)

    # Break the loop when 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release video capture
cap.release()
cv2.destroyAllWindows()
```
